Q1. What is statistics?

    Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, 
    presentation, and organization of data. It provides methods and techniques for summarizing and 
    describing data, making inferences and predictions from data, and testing hypotheses and theories 
    using statistical models. In applying statistics to a scientific, industrial, or social problem, 
    it is conventional to begin with statistical population or a statistical model to be studied. 
    Statistics is a mathematical body of science that pertains to the collection, analysis, 
    interpretation or explanation, and presentation of data. Statistics deals with how data can be 
    used to solve complex problems. Some people consider statistics to be a distinct mathematical 
    science rather than a branch of mathematics.

    Statistics can be divided into two main categories: descriptive statistics and inferential 
    statistics. Descriptive statistics involve methods for organizing and summarizing data, such 
    as measures of central tendency (mean, median, mode) and measures of variability (range, 
    standard deviation). Inferential statistics involve using sample data to make inferences 
    about a larger population, such as testing hypotheses and estimating population parameters.

Q2. Define the different types of statistics and give an example of when each type might be used.

    Statistics can be divided into two main types: 
    1.Descriptive Statistics: Discriptive statistics uses data that provides a description of the
        population either through numerical calculation or graph or table. It provides a graphical
        summary of data. It is simply used for summarizing objects, etc. 

    2. Inferential Statistics: Inferential Statistics makes inference and prediction about population
        Inferential Statistics makes inference and prediction about population based on a sample of 
        data taken from population. It generalizes a large dataset and applies probabilities to draw 
        a conclusion. It is simply used for explaining meaning of descriptive stats. It is simply 
        used to analyze, interpret result, and draw conclusion. Inferential Statistics is mainly 
        related to and associated with hypothesis testing whose main target is to reject null 
        hypothesis.

    Some example of each type: 
    Descriptive Statistics:
        Measures of central tendency: Mean, median, and mode. These measures describe the average 
        value of a dataset. For example, the mean height of a group of people.

        Measures of dispersion: Standard deviation, range, and variance. These measures describe 
        the spread or variability of a dataset. For example, the range of test scores for a group 
        of students.

        Frequency distributions: This describes the frequency or number of times each value in a 
        dataset appears. For example, a frequency distribution of the number of pets owned by a 
        group of people.

    Inferential Statistics: 
        Hypothesis testing: This is used to test whether a sample is likely to have come from a 
        specific population. For example, testing whether a new drug is effective by comparing it 
        to a control group.

        Confidence intervals: This is used to estimate the range of values that a population 
        parameter might take based on a sample. For example, estimating the average salary of a 
        profession based on a sample of data.

        Regression analysis: This is used to identify the relationship between two or more variables. 
        For example, analyzing the relationship between age and income.

Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

    Level of measurement also called scales of measurement, tells us how precisely variables
    are recorded. It refer to the different ways in which variables can be measured.
    There are four levels of measurement: nominal, ordinal, interval, and ratio.

    1. Nominal: Nominal variables are categorical variables that have no intrinsic order or 
        numerical value. They are simply classified into categories or groups. Examples of nominal 
        variables include gender, race, and religion.

    2. Ordinal: Ordinal variables have categories that can be ranked or ordered, but the difference 
        between the categories is not uniform. In other words, the intervals between each category 
        are not equal. Examples of ordinal variables include education level (e.g. elementary school, 
        high school, college), socioeconomic status (e.g. low, middle, high), and customer 
        satisfaction ratings (e.g. very satisfied, somewhat satisfied, not satisfied).

    3. Interval: Interval variables have categories that are ranked and have equal intervals 
        between them. However, they do not have a true zero point. This means that a value of 
        zero does not represent the absence of the variable being measured, but rather a point 
        on the scale. Examples of interval variables include temperature measured in degrees 
        Celsius or Fahrenheit, and IQ scores.

    4. Ratio: Ratio variables have categories that are ranked and have equal intervals between 
        them, and also have a true zero point. This means that a value of zero represents the 
        absence of the variable being measured. Examples of ratio variables include weight, height, 
        age, and income.

Q8. Which type of plot can be used to display data in terms of range?

    representation of the distribution of numerical data. To construct histogram, the 
    first step is to "bin"(or "bucket") the range of values - that is, divide the entire
    range of values into a series of intervals and then count how many values fall into 
    each interval. The bins are ususlly specified as consecutive, non-overlapping intervals
    of a variable. The bins must be adjcent and are often of equal size.

Q10. What are some common measures of central tendency and variability used in statistics? Explain
how each measure can be used to describe a dataset.

    Measures of central tendency and variability are statistical measures used to describe 
    and summarize a dataset. They help to give an idea of the typical values in the dataset
    and how spread out the data is.

    Common measures of central tendency include:

    Mean: The mean is the average value in a dataset. It is calculated by adding up all the 
    values in the dataset and dividing by the number of values. The mean is useful for describing 
    the typical value in a dataset.

    Median: The median is the middle value in a dataset when the values are arranged in order. 
    If there is an even number of values, the median is the average of the two middle values. 
    The median is useful for describing the central value in a dataset when there are outliers 
    or the data is skewed.

    Mode: The mode is the value that occurs most frequently in a dataset. It is useful for 
    describing the most common value in a dataset.

    Common measures of variability include:

    Range: The range is the difference between the highest and lowest values in a dataset. It 
    is useful for describing the spread of the dataset.

    Variance: The variance is a measure of how spread out the values in a dataset are from the mean. 
    It is calculated by taking the average of the squared differences between each value and the 
    mean. The variance is useful for describing the variability in a dataset.

    Standard deviation: The standard deviation is the square root of the variance. It is a measure 
    of how spread out the values in a dataset are from the mean. The standard deviation is useful 
    for describing the variability in a dataset and is often used in inferential statistics to 
    calculate confidence intervals and test statistics.
