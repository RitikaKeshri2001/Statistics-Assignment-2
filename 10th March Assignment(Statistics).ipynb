Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

    Estimation statistics, or simply estimation, is a data analysis framework 
    that uses a combination of effect sizes, confidence intervals, precision 
    planning, and meta-analysis to plan experiments, analyze data and interpret 
    results.

    Point estimate refers to a single numerical value that is used to estimate 
    an unknown population parameter. This value is calculated using a sample 
    statistic, such as the sample mean or sample proportion. For example, if 
    we want to estimate the mean salary of all employees in a company, we could 
    take a random sample of employees and calculate the sample mean salary.
    This sample mean would be the point estimate of the population mean salary.

    Interval estimate, on the other hand, provides a range of values within which 
    the true population parameter is likely to lie. This range is calculated based 
    on the sample data and is represented by a confidence interval. Confidence 
    intervals are usually expressed as a range of values with a specified level 
    of confidence. For example, if we construct a 95% confidence interval for the 
    population mean salary, we can say that we are 95% confident that the true 
    population mean salary falls within the interval.

Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.

```python
from math import sqrt
def population_mean(sample_mean, std, sample_size, z_score):
    standard_error = std/sqrt(sample_size)
    margin_error = z_score * standard_error
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    population_mean = (lower_bound + upper_bound)/2
    return population_mean
```

Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

    The purpose of statistical inference is to draw conclusions about a population 
    on the basis of data obtained from a sample of that population. 
    Hypothesis testing is a statistical method used to determine whether a particular 
    hypothesis about a population parameter is true or false. It involves the 
    formulation of two hypotheses, the null hypothesis and the alternative hypothesis, 
    and then using sample data to determine which hypothesis is supported by the 
    evidence.The null hypothesis (H0) is the default hypothesis, which assumes that 
    there is no significant difference or relationship between the population parameter 
    and the sample statistic. The alternative hypothesis (Ha) is the hypothesis that 
    contradicts the null hypothesis and suggests that there is a significant difference 
    or relationship between the population parameter and the sample statistic.

    Hypothesis testing is used to make decisions based on the evidence from the sample 
    data. It is commonly used in scientific research, quality control, business analysis,
    and many other fields to test theories and make data-driven decisions.

    The importance of hypothesis testing lies in its ability to provide a structured 
    and objective way to evaluate the validity of a hypothesis. It helps to avoid making 
    incorrect conclusions based on random or biased data. Hypothesis testing also helps 
    to increase the credibility and reproducibility of research findings by providing a 
    transparent and systematic way to evaluate the evidence. Additionally, hypothesis 
    testing allows researchers to compare the effectiveness of different interventions 
    or treatments and identify factors that may affect the outcome.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

```python
import numpy as np
from scipy.stats import ttest_ind

# Define the two samples
sample1 = np.array([1, 2, 3, 4, 5])
sample2 = np.array([6, 7, 8, 9, 10])

# Calculate the means and standard deviations of the samples
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the degrees of freedom
df = len(sample1) + len(sample2) - 2

# Calculate the pooled standard deviation
s_p = np.sqrt(((len(sample1) - 1) * std1 ** 2 + (len(sample2) - 1) * std2 ** 2) / df)

# Calculate the t-statistic
t = (mean1 - mean2) / (s_p * np.sqrt(1 / len(sample1) + 1 / len(sample2)))

# Calculate the p-value
p = 2 * (1 - ttest_ind(sample1, sample2)[1])

# Print the results
print("Sample 1 mean:", mean1)
print("Sample 2 mean:", mean2)
print("Sample 1 standard deviation:", std1)
print("Sample 2 standard deviation:", std2)
print("Degrees of freedom:", df)
print("Pooled standard deviation:", s_p)
print("T-statistic:", t)
print("P-value:", p)
```

Q7: Write down the steps involved in hypothesis testing.

    The following are the general steps involved in hypothesis testing:

    1. State the null and alternative hypotheses: The first step is to clearly 
    state the null and alternative hypotheses. The null hypothesis assumes that 
    there is no relationship between variables or no significant difference between 
    groups. The alternative hypothesis assumes that there is a significant 
    relationship or difference.

    2. Determine the level of significance: The level of significance determines 
    how much evidence is required to reject the null hypothesis. 

    3. Choose a test statistic: The test statistic is used to calculate the probability 
    of obtaining the observed data if the null hypothesis is true. The choice of test 
    statistic depends on the type of data being analyzed and the hypothesis being tested.

    4. Calculate the p-value: The p-value is the probability of obtaining a result as 
    extreme as, or more extreme than, the observed data if the null hypothesis is true. 
    If the p-value is less than or equal to the level of significance, then the null 
    hypothesis is rejected.

    5. Interpret the results: If the null hypothesis is rejected, then the alternative 
    hypothesis is accepted. If the null hypothesis is not rejected, then no conclusion 
    can be made about the alternative hypothesis.

    6. Draw conclusions: Based on the results, conclusions are drawn about the hypothesis 
    being tested. These conclusions should be based on the level of significance, the 
    p-value, and the practical significance of the findings.

Q8. Define p-value and explain its significance in hypothesis testing.

    The p-value is a statistical measure that indicates the probability of 
    obtaining an observed result (or one more extreme) if the null hypothesis 
    is true. In other words, it is the probability of getting a result as extreme 
    as the one observed, assuming that the null hypothesis is true.

    In hypothesis testing, the null hypothesis is a statement that there is no 
    significant difference between two populations, while the alternative 
    hypothesis is a statement that there is a significant difference between 
    two populations. The p-value is used to determine the statistical significance 
    of the observed data and to help decide whether to reject or fail to reject 
    the null hypothesis.

    The significance level, also known as the alpha level, is a pre-determined 
    threshold that is used to evaluate the p-value. If the p-value is less than 
    the significance level (i.e., p < alpha), the null hypothesis is rejected, 
    and it is concluded that there is evidence to support the alternative 
    hypothesis. On the other hand, if the p-value is greater than or equal to 
    the significance level (i.e., p >= alpha), the null hypothesis is not rejected, 
    and it is concluded that there is insufficient evidence to support the 
    alternative hypothesis.

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stat

# degree of freedom
df = 10

# create an array of x-values from -4 to 4
x = np.linspace(-4, 4, 1000)

# calculate student's t-distribution object
t_dist = stat.t(df)

# Calculate the corresponding y-values for each x-value
y = t_dist.pdf(x)

plt.plot(x, y)
plt.show()
```

Q11: What is Student’s t distribution? When to use the t-Distribution.

    Student’s t-distribution is a probability distribution that is used to 
    estimate the mean of a population when the sample size is small, or the 
    population standard deviation is unknown. It is named after the pseudonym 
    used by William Sealy Gosset, a British statistician who developed the 
    distribution in 1908 while working at Guinness Brewery.

    The t-distribution is similar to the standard normal distribution, with 
    the main difference being that it has fatter tails. The degree of freedom 
    (df) parameter determines the shape of the t-distribution. The degrees of 
    freedom are equal to the sample size minus one, which means that as the sample 
    size increases, the t-distribution approaches the standard normal distribution.

    The t-distribution is commonly used in statistical inference, particularly in 
    hypothesis testing and confidence interval estimation. It is used when the 
    sample size is small, typically less than 30, or when the population standard 
    deviation is unknown.

    When the sample size is large, the sample mean follows the normal distribution. 
    However, when the sample size is small, the sample mean follows the t-distribution. 
    Therefore, when the sample size is small, the t-distribution should be used to make 
    inferences about the population mean.

Q12: What is t-statistic? State the formula for t-statistic.

    The t-statistic is a measure of how far the sample mean deviates from the null 
    hypothesis population mean in units of the standard error. It is used in 
    hypothesis testing and is based on the t-distribution.

    The formula for t-statistic is:

    t = (x̄ - μ) / (s / √n)

    where:
    x̄ is the sample mean
    μ is the null hypothesis population mean
    s is the sample standard deviation
    n is the sample size

    The t-statistic is calculated by subtracting the null hypothesis population 
    mean from the sample mean and dividing the result by the standard error. The 
    standard error represents the variability of the sample mean and is calculated 
    by dividing the sample standard deviation by the square root of the sample size.

    The t-statistic measures the difference between the sample mean and the null 
    hypothesis population mean in standard deviation units. If the t-statistic is 
    large, it indicates that the sample mean is significantly different from the 
    null hypothesis population mean, and we can reject the null hypothesis.

    In hypothesis testing, the t-statistic is compared to a critical value from the 
    t-distribution, which depends on the sample size and the significance level of 
    the test. If the t-statistic exceeds the critical value, we reject the null 
    hypothesis and accept the alternative hypothesis. Otherwise, we fail to reject 
    the null hypothesis.

sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.

    To estimate the population mean revenue with a 95% confidence interval, we 
    can use the following formula:
    Confidence Interval = x̄ ± z*(s/√n)

    where x̄ is the sample mean revenue, s is the sample standard deviation, n 
    is the sample size, and z is the z-score associated with the desired level 
    of confidence. For a 95% confidence interval, the z-score is 1.96.

    Sample size (n) = 50
    Sample mean (x̄) = $500
    Sample standard deviation (s) = $50
    Level of confidence = 95%

    confidence interval = 500 ± 1.96*($50/√50) 
                        = 500 + 14.14 , 500 - 14.14
                        = 514.14 , 485.86

    Therefore, the 95% of confidence interval for the population mean
    is $485.86 to $514.14.

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

    To test the hypothesis with a significance level of 0.05, we can use 
    a one-sample t-test.

    Sample size (n) = 100
    population mean (μ) = 10 mmHg
    Sample mean (x̄) = 8 mmHg
    Sample standard deviation (s) = 3 mmHg
    significance value (α) = 0.05
    confidence interval = 95% or 0.95

    Step 1: Determine the null hypothesis and alternate hypothesis
    Null hypothesis(H0) : μ = 10
    Alternate hypothesis (H1) : μ < 10 (one tail)

    Step 2: Determine the significance level, which is 0.05

    Step 3: calculate the t-test
    t = (8 - 10) / (3 / √100) = -6.67

    Step 4: Determine the critical value
    Degrees of freedom = 100 - 1 = 99
    Critical value for a one-tailed test with 99 degrees of freedom and a 
    significance level of 0.05 is -1.66.

    step 5: compare the t-test(-6.67) to critical value(-1.66)
    Since the test statistic (-6.67) is less than the critical value (-1.66), we 
    reject the null hypothesis. Therefore, we can conclude that the new drug 
    significantly decreases blood pressure by more than 10 mmHg.

Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.

    To test the hypothesis that the population means for the two groups are 
    equal with a significance level of 0.01, we can use a two-sample t-test. 

    sample size (n): n1 = 30, n2 = 40
    Sample mean (x̄): x̄1 = 80, x̄2 = 75
    Sample standard deviation (s): s1 = 10, s2 = 8

    Step 1: State null hypothesis and alternate hypothesis
    Null hypothesis(H0) : Population means for the two groups are equal. (x̄1 = x̄2)
    Alternate hypothesis (H1) : population means for the two groups are not 
                                equal. (x̄1 != x̄2) (Two tail)

    Step 2: Determine level of significance, which is 0.01  
        Calculate the pooled standard deviation
        The pooled standard deviation is 
        s_p = √(((n1 - 1) * s1^2 + (n2 - 1) * s2^2) / (n1 + n2 - 2))
            = √(((30-1) * 10^2 + (40-1) * 8^2)/(30 + 40 - 2))
            = 9.14

    Step 3: calculate the t-test
    t = (x̄1 - x̄2) / (s_p * √(1/n1 + 1/n2))
      = (80 - 75) / (9.14 * √(1/30 + 1/40)) 
      = 2.29

    Step 4: Determine the critical value
    degree of freedom:
    df = n1 + n2 - 2 = 30 + 40 - 2 = 68
    The critical value for a two-tailed test with 68 degrees of freedom and 
    a significance level of 0.01 is approximately ±2.39.

    Step 5: compare the t-test to critical value
    Since the test statistic (2.29) is not outside the critical region, we 
    fail to reject the null hypothesis. Therefore, the population means for 
    the two groups are equal.

    can use the following formula:
    CI = X̄ ± Z* (σ / √n)

    where CI is the confidence interval, X̄ is the sample mean, Z* 
    is the critical value.

    sample mean = 4
    sample size = 50
    sample standard deviation = 1.5 

    The critical value for a 99% confidence level is approximately 2.58.

    CI = 4 ± 2.58 * (1.5 / √50) = 3.58, 4.42

    Therefore, we can estimate with 99% confidence that the population 
    mean number of ads watched by viewers during a TV program is between 
    3.58 and 4.42.
