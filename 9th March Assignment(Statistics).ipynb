Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

    Probability Mass Function (PMF) and Probability Density Function (PDF) are two important 
    concepts in probability theory and statistics. They are used to describe the probability 
    distribution of a random variable.

    Probability Mass Function(PMF): 
        In probability and statistics, a probability mass function is a function that gives 
        the probability that a discrete random variable is exactly equal to some value.Sometimes
        it is also known as the discrete density function. The pmf is often the primary means 
        of defining a discrete probability distribution, and such functions exist for either
        scalar or multivariate random variables whose domain is discrete.
        It maps each value of the random variable to a probability. The sum of probabilities 
        for all possible values of the random variable must equal 1.
        Let X be the discrete random variable. Then the formula for the probability mass 
        function, f(x), evaluated at x, is given as follows:

        f(x) = P(X = x)

        For example, 
        Rolling a dice, {1,2,3,4,5,6}
        let's consider a fair six-sided dice. The PMF for this dice would give the probability 
        of each possible outcome of rolling the dice, which is 1/6 for each value from 1 to 6.

    Probability Distribution Function(PDF):
        In probability theory and statistics, a probability distribution is the mathematical
        function that gives the probabilities of occurrence of different possible outcomes 
        for an experiment. Or, a Probability Density Function (PDF) is a function that gives 
        the probability of a continuous random variable falling within a particular range of 
        values. It describes the distribution of the random variable as a continuous function. 
        The area under the curve of the PDF for a specific range of values represents the 
        probability of the variable falling within that range.

        For example, 
        let's consider the height of adult males in a population. The PDF for this variable would 
        describe the distribution of heights across the population, and give the probability of 
        a person's height falling within a particular range of values. The PDF for height would 
        likely be a bell-shaped curve with the highest probability density occurring near the 
        mean height for the population.

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

    Cumulative Density Function (CDF) is a mathematical function that describes the 
    probability that a random variable X takes on a value less than or equal to a 
    specific value x. It is a function that maps any value of x to the probability 
    that X takes on a value less than or equal to that value x.
    The CDF of a random variable X is defined as:

    F(x) = P(X ≤ x)
    where F(x) is the cumulative distribution function and P(X ≤ x) is the probability 
    that X takes on a value less than or equal to x.

    An example of a CDF is the cumulative distribution function for a standard normal 
    distribution. In this case, the CDF gives the probability that a standard normal 
    random variable Z takes on a value less than or equal to a given value of z. This 
    function is denoted as Φ(z) and is a continuous and increasing function.

    The CDF is useful because it provides a complete description of the probability 
    distribution of a random variable. It can be used to calculate the probability of 
    X taking on any specific value, as well as the probability of X falling within a 
    specific range of values. The CDF can also be used to calculate statistical 
    measures such as the mean, variance, and standard deviation of a random variable.

    Another advantage of the CDF is that it allows for the comparison of different 
    probability distributions. By comparing the CDFs of two different random variables,
    we can determine which variable has a higher probability of taking on a specific 
    value or falling within a specific range of values.

Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.

    The normal distribution, also known as the Gaussian distribution, is a common 
    probability distribution used to model a variety of phenomena in different 
    fields such as science, finance, engineering, and more. Some examples of 
    situations where the normal distribution might be used as a model are:

    1. Physical measurements: The normal distribution can be used to model errors 
        or deviations from a mean value in physical measurements such as length, 
        weight, or temperature.
    2. IQ scores: IQ scores are typically normally distributed, with a mean of 
        100 and a standard deviation of 15
    3. Financial data: The normal distribution can be used to model financial 
        data, such as stock prices or returns, as well as interest rates and bond 
        yields.
    4. Medical data: The normal distribution is often used to model the distribution 
        of medical data such as blood pressure, cholesterol levels, or heart rate.

    The normal distribution is characterized by two parameters, the mean (μ) and 
    the standard deviation (σ). The mean represents the center of the distribution, 
    and the standard deviation represents the spread of the distribution. The shape 
    of the normal distribution is symmetrical around the mean, with the majority of 
    the data falling within one, two, or three standard deviations from the mean.

    The mean determines the location of the peak of the distribution, while the 
    standard deviation determines the width of the distribution. If the standard 
    deviation is large, the distribution will be wider, indicating a greater 
    variability in the data. Conversely, if the standard deviation is small, the 
    distribution will be narrower, indicating less variability in the data.

    Bernoulli distribution is a discrete probability distribution that 
    describes the probability of a single binary event occurring. It is 
    named after Jacob Bernoulli, a Swiss mathematician who first introduced 
    the concept in the 17th century. It can be thought of as a model for 
    the set of possible outcomes of any single experiment that asks a yes–no 
    question. Such questions lead to outcomes that are boolean-valued. In a 
    Bernoulli distribution, there are only two possible outcomes for an event: 
    success (usually represented by 1) or failure (usually represented by 0). 
    The probability of success is denoted by the parameter p, while the 
    probability of failure is denoted by q (1-p).

    An example of Bernoulli distribution could be tossing a fair coin, where 
    success is defined as getting a head and failure is defined as getting a tail. 
    The probability of getting a head (success) is 0.5, while the probability of 
    getting a tail (failure) is also 0.5.

    The key difference between Bernoulli distribution and binomial distribution is 
    that Bernoulli distribution describes the probability of a single binary event 
    occurring, while binomial distribution describes the probability of multiple 
    binary events occurring.

Q7: Explain uniform Distribution with an example.

    Uniform distribution is a probability distribution in which every 
    possible outcome is equally likely to occur. In other words, a uniform 
    distribution occurs when each value within a certain range has an equal 
    chance of being selected.

    For example, suppose we want to select a random number between 1 and 10 
    (inclusive), and we assume that every number within that range is equally 
    likely to be chosen. In this case, we have a uniform distribution with a 
    range of 1 to 10.

    To calculate the probability of selecting any specific number within that 
    range, we divide the number of possible outcomes (which is 10) by the range 
    of possible values (which is also 10). Therefore, the probability of selecting 
    any one number is 1/10, or 0.1.

    In a uniform distribution, the probability density function is a horizontal line. 
    This means that the probability of selecting any number within the range is 
    constant and does not vary. In the case of our example, the probability density 
    function would be a horizontal line at a height of 0.1 from x=1 to x=10.

Q8: What is the z score? State the importance of the z score.

    A z-score, also known as a standard score, is a measure of how many 
    standard deviations an observation or data point is away from the mean 
    of its distribution. It is calculated by subtracting the mean from the 
    observation and dividing by the standard deviation.
    The formula for calculating the z-score for an observation x is:
    z = (x - μ) / σ
    where μ is the mean of the distribution and σ is the standard deviation.

    The importance of the z-score lies in its usefulness in comparing observations 
    from different distributions. Since the z-score is a standardized value, it
    allows us to compare observations from different distributions, regardless of 
    the units of measurement or the scales of the distributions.
    Another important use of the z-score is in hypothesis testing and statistical 
    inference. By converting observations to z-scores, we can use standard normal
    distribution tables to calculate probabilities and test hypotheses about the 
    population from which the sample was drawn. The z-score is also used in quality 
    control and process improvement to identify outliers and assess the performance 
    of a process.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

    The Central Limit Theorem (CLT) is a fundamental theorem in statistics 
    that describes the behavior of the sample means of a population. It states 
    that, under certain conditions, the sample means from a sufficiently large 
    number of independent and identically distributed random variables will be 
    normally distributed, regardless of the underlying distribution of the 
    population from which the samples are drawn.

    The significance of the Central Limit Theorem lies in its applicability to 
    a wide range of statistical analyses. It allows us to make inferences about 
    a population based on a sample, even if we do not know the underlying distribution 
    of the population. It also enables us to estimate confidence intervals and 
    perform hypothesis testing, which are key tools in statistical inference.

    Furthermore, the Central Limit Theorem is used in many practical applications, 
    such as quality control, finance, and engineering, where it is important to 
    estimate the parameters of a population from a sample. It also provides a 
    theoretical justification for the use of many statistical techniques, such 
    as the t-test and the analysis of variance (ANOVA).

    Assumption of central limit theorem are:-

    3. Finite variance: The population from which the sample is drawn has a

    distributed.
