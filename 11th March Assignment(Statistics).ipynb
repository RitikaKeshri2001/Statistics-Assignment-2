Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

    In hypothesis testing, A type I error (false-positive) occurs if an 
    investigator rejects a null hypothesis that is actually true in the 
    population, A type II error (false-negative) occurs if the investigator 
    fails to reject a null hypothesis that is actually false in the population.

    Here are some example scenarios for each type of error:

    Type 1 error: Suppose a drug company is testing a new medication for a disease. 
    The null hypothesis is that the medication has no effect on the disease. 
    The alternative hypothesis is that the medication is effective in treating the 
    disease. The drug company conducts a study and concludes that the medication 
    is effective, and they begin marketing it. However, later research shows that 
    the medication does not actually have an effect on the disease. This is a 
    Type 1 error because the null hypothesis was rejected (i.e., the medication was 
    deemed effective) when it was actually true (i.e., the medication had no effect).

    Type 2 error: Suppose a school district is testing a new teaching method for 
    improving math scores. The null hypothesis is that the teaching method has no 
    effect on math scores. The alternative hypothesis is that the teaching method 
    is effective in improving math scores. The school district conducts a study but 
    does not find a significant difference in math scores between the group that 
    received the new teaching method and the control group. They conclude that the 
    teaching method is not effective and do not implement it. However, later research 
    shows that the teaching method is effective. This is a Type 2 error because the 
    null hypothesis was not rejected (i.e., the teaching method was deemed ineffective) 
    when it was actually false (i.e., the teaching method was effective).

Q4: Explain Bayes's theorem with an example.

    Bayes's theorem(alternatively Bayes' law or Bayes' rule), named after Thomas 
    Bayes,is a mathematical formula that describes the probability of an event 
    occurring based on prior knowledge of conditions that might be related to the 
    event. It is often used in statistics and machine learning to update 
    probabilities as new information becomes available.
    
    For example, if the risk of developing health problems is known to increase with 
    age, Bayes' theorem allows the risk to an individual of a known age to be assessed 
    more accurately by conditioning it relative to their age, rather than simply 
    assuming that the individual is typical of the population as a whole.

    The formula for Bayes's theorem is as follows:

    P(A|B) = P(B|A) x P(A) / P(B)

    Where:
    P(A|B) is the probability of event A given that event B has occurred.
    P(B|A) is the probability of event B given that event A has occurred.
    P(A) is the prior probability of event A occurring.
    P(B) is the prior probability of event B occurring.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

    A confidence interval (CI) is a range of estimates for an unknown 
    parameter. A confidence interval is a range of values that is used to 
    estimate the unknown true value of a population parameter with a certain 
    level of confidence. In other words, it's a range of values that we 
    believe the true value of a parameter lies within with a certain degree 
    of certainty.

    For example, suppose we want to estimate the average height of all adult 
    males in a particular city. We take a sample of 100 adult males from the 
    city and measure their height. The sample mean height is 175 cm, and the 
    sample standard deviation is 6 cm. We want to construct a 95% confidence 
    interval for the true population mean height.

    To calculate the confidence interval, we can use the formula:

    Confidence interval = sample mean ± (critical value) x (standard error)
                        = x̄ ± Z * (σ/√n)

    For a 95% confidence level and a sample size of 100, the critical value is
    approximately 1.96 

    standard error = 6 / sqrt(100)
                   = 0.6
    
    critical value(Z) = 1.96

    Confidence interval = 175 ± 1.96 x 0.6
                        = (173.44, 176.56)

    Therefore, we can say with 95% confidence that the true population mean 
    height lies within the interval of 173.44 cm and 176.56 cm.

Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.

    Bayes' Theorem is a formula used to calculate the probability of an event 
    occurring, given prior knowledge of the event's probability and new evidence. 
    It is written as follows:

    P(A|B) = P(B|A) * P(A) / P(B)

    where P(A|B) is the probability of A given B, 
    P(B|A) is the probability of B given A, 
    P(A) is the prior probability of A, and 
    P(B) is the prior probability of B.

    Example:
    Suppose a certain city has two taxi companies, A and B. It is known that 60% 
    of the taxis in the city are from company A, while the remaining 40% are from 
    company B. Company A's taxis are involved in 4% of all accidents in the city, 
    while company B's taxis are involved in 2% of all accidents. If a taxi is 
    involved in an accident, what is the probability that it is from company A?

    Solution:- 
    We want to calculate the probability of a taxi being from company A given that 
    it is involved in an accident. Using Bayes' Theorem, we have:

    P(A|B) = (P(B|A) * P(A)) / P(B)

    To calculate P(B), we need to first calculate P(B|not A), which is the probability 
    of an accident given that the taxi is not from company A.
    P(B|not A) = 0.02

    since company B's taxis are involved in 2% of all accidents. The prior probability 
    of a taxi not being from company A is:
    P(not A) = 1-P(A) = 0.40

    P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
         = 0.04 * 0.60 + 0.02 * 0.40
         = 0.028

    Substituting these values into Bayes' Theorem, we get:

    P(A|B) = (0.04 * 0.60) / 0.028
           = 0.857

    Therefore, the probability of a taxi being from company A given that it is 
    involved in an accident is 0.857 or approximately 86%.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

    The margin of error is a statistical term that measures the range of 
    uncertainty in a confidence interval. It is the amount by which the 
    sample estimate may differ from the true population parameter, and it 
    is typically expressed as a percentage.

    Margin of error is affected by sample size as:
    The larger the sample size, the smaller the margin of error.This is because 
    larger sample sizes provide more information and reduce the impact of 
    random sampling errors.

    For example, suppose a pollster wants to estimate the proportion of adults in 
    a city who support a proposed tax increase. They take a random sample of 400 
    adults and find that 60% of them support the tax increase. Using a 95% confidence 
    interval, they calculate a margin of error of ±5%. This means that they are 95% 
    confident that the true proportion of adults who support the tax increase falls 
    within the range of 55% to 65%.

Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.

    The z-score measures the distance between a data point and the mean of its 
    distribution in units of standard deviation. It is calculated as:

    z = (x - μ) / σ
    where, x is the data point,
    μ is the population mean,
    σ is the population standard deviation

    z = (75 - 70)/5 = 1

    Interpretation: The z-score of 1 indicates that the data point is one 
    standard deviation above the mean of its distribution. This means that 
    the data point is relatively far from the center of the distribution and 
    is considered to be a moderately extreme value. The positive sign of the 
    z-score indicates that the data point is above the mean.

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

    To conduct a hypothesis test to determine if the drug is significantly 
    effective at a 95% confidence level using a t-test, 
    t = (x̄ - μ) / (s / √n)

    where x̄ is the sample mean, μ is the population mean, 
    s is the sample standard deviation, and n is the sample size.

    Sample mean (x̄) = 6 pounds
    Sample standard deviation (s) = 0.5
    confidence interval = 95%
    significance value (α) = 0.05

    Step 1: To determine null hypothesis and alternate hypothesis
    Null hypothesis (H0): The new weight loss drug is not significantly effective 
                            (mean weight loss = 0).
    Alternative hypothesis (H1): The new weight loss drug is significantly effective 
                            (mean weight loss != 0) (two-tailed)
    
    Step 2: Determine level of significance, which is 0.05

    Step 3: calculate the t-test
    t = (6 - 0) / (2.5 / √50) = 12

    step 4: Determine the critical value
    Degrees of freedom = 50 - 1 = 49
    The critical value of the t-distribution with 49 degrees of freedom and a 
    significance level of 0.05 is 2.009

    step 5: compare the t-test to critical value
    Since the test statistic is greater than the critical value, we reject the 
    null hypothesis and conclude that the drug is significantly effective at a 
    95% confidence level.
